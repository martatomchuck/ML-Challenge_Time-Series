```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

# from sklearn.metrics import mean_absolute_error as mae

import matplotlib.pyplot as plt
%matplotlib inline
```

Przygotowanie danych

```python
path = 'https://raw.githubusercontent.com/dataworkshop/5dwchallenge_2019/master/challenge5/input/mauna_loa.csv'
df = pd.read_csv(path)

df.head()
 
```

```python
df['date'] = df.apply( lambda row: '{}-{}-{}'.format(int(row['year']), int(row['month']), int(row['day'])), axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['ppm_fixed'] = df['ppm'].map(lambda x: np.nan if x < 0 else x).fillna(method='backfill')
df[['ppm_fixed']].plot();
```

# PROGNOZOWANIE

```python
#trenowanie danych na jednym podzbiorze, a prognozowanie na innym
#potem: porównanie obu zbiorów

cut_year = 2010
train = df[df.year < cut_year]
test = df[df.year >= cut_year]

plt.figure(figsize=(15,10))

plt.plot(train.date, train.ppm_fixed, label='train');
plt.plot(test.date, test.ppm_fixed, label='test');

plt.legend();
```

# Prophet

```python
#model oczekuje na dwie kolumny:
#w pierwszej jest oś czasu (ds), a w drugiej jest oś wartości (y)

fb_df = train[['date', 'ppm_fixed']].copy()
fb_df.columns = ['ds', 'y']

fb_df.head()
```

```python
#TRENOWANIE MODELU
m = Prophet()
m.fit(fb_df)
```

```python
len(test)
```

```python
#periods = długość zestawu testowego
#freq = tygodniowe ('W')
#include_history - bez danych historycznych, testowych
future = m.make_future_dataframe(periods=len(test), freq='W', include_history=False)

future.head()
future.tail()
```

```python
test.tail()
```

# PROGNOZA

```python
forecast = m.predict(future)
forecast.head()

yhat = WARTOŚĆ PROGNOZOWANA (najbardziej nas interesuje)
```

# WIZUALIZACJA PROGNOZY

```python
plt.figure(figsize=(15,15))

plt.plot(train.date, train.ppm_fixed, label='train');
plt.plot(test.date, test.ppm_fixed, label='test');
plt.plot(test.date, forecast.yhat, label='forecast');

plt.legend();
```

```python
#wizualizacja samej prognozy, bez danych testowych
plt.figure(figsize=(15,5))

plt.plot(test.date, test.ppm_fixed, label='test');
plt.plot(test.date, forecast.yhat, label='forecast');

plt.legend();
```
